`df_bikes.isna().any` gathers any and all null values.

Correcting null values depends on the column and dataset. Some useful strategies are:
1. **Replacing with the median/mean**
    - The idea here is to replace null values with the average column value.
    - To make changes permanent set `inplace=True` in `fillna()` method    

```python
df_bikes.windspeed.fillna((df_bikes.windspeed.median()), inplace=True)
df_bikes.iloc[[56, 81, 128]] # rows 56, 81 and 128 were null for windpeed
```

*Tip: `.iloc` uses single brackets for one index as follows: `df_bikes.iloc[56]`. **df_bikes** also accepts list inside brackets to allow multiple indices. Multiple indices require double brackets: `df_bikes.iloc[[56, 81]]]`*

2. **Groupby with the median/mean**
    
    A groupby organizes rows by shared values.
    - These values are coming from different rows, so we need a way to combine, or aggregate, the values. Choices may include `.sum()`, `.count()`, `.mean()` and `.median()`.

    Lets correct the null values in the **hum** column, short for **humidity**, we can take the median humidity by season.

```python
df_bikes.groupby(['season']).median() # Produced column values are medians
```

```python
df_bikes["hum"] = df_bikes["hum"].fillna(df_bikes.groupby("season")["hum"].transform('median'))
```

```python
df_bikes.iloc[[129, 213, 388]]
```

3. **Obtaining the median/mean from specific rows**
    
    In some cases, it may be advantageous to replace null values with data from specific rows.

    When correcting *temperature*, aside from consulting historical records, taking the mean temperature of the day before and the day after should give a good estimate.

```python
df_bikes[df_bikes.temp.isna()]
```

```python
mean_temp = (df_bikes.iloc[700]["temp"] + df_bikes.iloc[700]["temp"])/2
mean_atemp = (df_bikes.iloc[700]["atemp"] + df_bikes.iloc[700]["atemp"])/2

df_bikes.temp.fillna((mean_temp), inplace=True)
df_bikes.atemp.fillna((mean_atemp), inplace=True)

df_bikes.iloc[701]
```

4. Extrapolate Dates

    In order to correct null values involves dates, we have to first make sure that real dates are provided in the dataset. If so, date values may be extrapolated.

    `df_bikes["dteday"]` is a date column; however the type of the column is an object, commonly represented as a string. It may be converted to a **datetime**.

```python
import datetime as dt

df_bikes["dteday"] = pd.to_datetime(df_bikes.dteday, infer_datetime_format=True)

df_bikes["mnth"] = df_bikes["dteday"].dt.month
df_bikes.tail()
```

As you can see, the month values are all correct, but the year value needs to be changed. Year column **yr** is normalized, meaning it's converted to values between 0 and 1. 
- Normalized data is often more efficient because machine learning weights do not have to adjust for different ranges.

```python
df_bikes.loc[730, 'yr'] = 1.0
```

For machine learning, all data columns should be numerical. According to `.info()` method, the only column that is not numerical is `df_bikes["dteday"]`. Furthermore, it's redundant since all date information exists in other columns.

## Predicting Regression

In the bike rentals dataset, `df_bikes["cnt"]` is the number of bike rentals in a given day. Predicting this column would be of great use to a bike rental company.

According to the dataset, `df_bikes["cnt"]` is the sum of `df_bikes["casual"]` and `df_bikes["registered"]`. Thus, it makes no sense to include these input columns that would be unknown in reality.

All current columns may be used to predict bike count except for abovementioned columns.

```python
df_bikes = df_bikes.drop(["casual", "registered"], axis=1)
```

Now the dataset is cleaned and ready for machine learning task. It is best to save this form of dataset to file before going any further.

Afterwards, train and test sets will be created to be fed to ML pipeline.

```python
# index = False parameter prevents an additional col from being created by the index
df_bikes.to_csv("bike_rentals_cleaned_170222.csv", index=False)
```

## Linear Regression

First model to be inspected is **LinearRegression** model. Linear Regression takes input columns as polynomial variables, and multiplies the values with coefficients (weights) to predict the target column. **Gradient Descent** works under the hood to minimize the error. The predictions of linear regression could be any real number.

```python
from sklearn.model_selection import train_test_split
import warnings

warnings.filterwarnings('ignore')

X = df_bikes.iloc[:, :-1] # first : -> all the rows. second : -> from 1st col to the end without including it
y = df_bikes.iloc[:, -1]  # first : -> all the rows. second : -> only the last column

X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=2)
```

```python
from sklearn.linear_model import LinearRegression

lin_reg = LinearRegression()

lin_reg.fit(X_train, y_train)

y_pred = lin_reg.predict(X_test)
```

Now compare the predictions with the test set. Scoring the model requires a basis of comparison. The standard for linear regression is the **root mean squared error (RMSE)**.
- The RMSE requires two pieces: *mean_squared_error*, the sum of the squares of differences between predicted and actual values, and *square root* to keep the units the same.

```python
from sklearn.metrics import mean_squared_error
import numpy as np

mse = mean_squared_error(y_test, y_pred)
rmse = np.sqrt(mse)

print("RMSE: %0.2f" % rmse)
```

It is hard to know whether an error of **898** is good or bad without knowing the expected range of rentals per day.

RMSE has the same unit as dependent variable.

One important question to ask is *What is a good RMSE value?*. The short answer: *It depends on the range of the dataset*. The lower the RMSE, the better a given model is able to fit a dataset.

One way to gain a better understanding of whether a certain RMSE value is **good** is to normalize it using the formula:

`NormalizedRMSE = RMSE / (max - min)`

This produces a value between 0 and 1, where values closer to 0 represent better fitting models.

```python
df_bikes.cnt.describe()
```

```python
NormalizedRMSE = 898.10 / (8714 - 22)
NormalizedRMSE
```

## XGBoost Regressor

```python
from xgboost import XGBRegressor

xg_reg = XGBRegressor()
xg_reg.fit(X_train, y_train)

y_pred_xg = xg_reg.predict(X_test)

rmse_xg = np.sqrt(mean_squared_error(y_test, y_pred_xg))
print("RMSE: %0.2f" % rmse_xg)
```

```python
NormalizedRMSE_XGB = 691.98 / (8714 - 22)
NormalizedRMSE_XGB
```

```python
# Column headings are the entires of first row, so load dataset with header=None
df_census = pd.read_csv('https://archive.ics.uci.edu/ml/machine-learning-databases/adult/adult.data', header=None)
df_census.head()
```

```python
# More info https://archive.ics.uci.edu/ml/datasets/Census+Income
df_census.columns=['age', 'workclass', 'fnlwgt', 'education', 'education-num', 'marital-status', 'occupation', 'relationship', 'race', 'sex', 'capital-gain', 'capital-loss', 'hours-per-week', 'native-country', 'income']
df_census.head()
```

### Step 2: Clean up Null Values

We can verify that there is no null values in the dataset.

```python
df_census.info()
```

```python
df_census.isna().sum().sum()
```

### Step 3: Drop Columns that have no information value

A quick review of the *df_census* data reveals that **education_num** column is a numerical conversion of **education** column. Since the information is the same we may delete the non-numeric one.

```python
df_census = df_census.drop(["education"], axis=1)
```

### Step 4: Clean up Non-numerical Columns

All columns of the **dtype** object must be transformed into numerical columns. Pandas `get_dummies` method takes the non-numerical unique values of every column and converts them into their own column, with **1** indicating presence and **0** indicating absence. 

```python
df_census = pd.get_dummies(df_census)
df_census.head()
```

The target column is whether or not someone makes 50K. After `pd.get_dummies`, two columns **income_ <=50K** and **income_ >50K** are used to determine whether someone makes 50K. Since either column will work, we may delete one of them.

```python
df_census = df_census.drop(["income_ <=50K"], axis=1)
```

### Split the data into X (predictor) and y (target) columns

```python
X = df_census.iloc[:, :-1]
y = df_census.iloc[:, -1]

X.shape
```

### Apply Classification with Logistic Regression

Mathematically, logistic regression works in a manner similar to linear regression. For each column, logistic regression finds an appropriate weight that maximizes the model accuracy. The primary difference is that instead of summing each term, logistic regression uses the **sigmoid function**.

Let's use cross-validation on logistic regression to predict whether someone makes over 50K.

```python
from sklearn.model_selection import cross_val_score

def cross_val(classifier, num_splits=10):
    model = classifier
    scores = cross_val_score(model, X, y, cv=num_splits)

    print(f"Accuracy: {np.round(scores, 2)}")
    print(f"Accuracy mean: {scores.mean()}")
```

```python
from sklearn.linear_model import LogisticRegression
cross_val(LogisticRegression())
```

### Try with XGBoost Classifier

```python
from xgboost import XGBClassifier
cross_val(XGBClassifier(n_estimators=5))
```
